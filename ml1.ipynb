```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_pickle('citation-ml.pkl')
```

```python
ax = df[df.n_citation < 100].n_citation.hist(bins=30, figsize=(8,5))
ax.set_ylabel('Frequency')
ax.set_title('Number of Citation per paper')
ax.set_xlabel('Number of Citation')
```

```python
ax = df.groupby('year').mean().plot(kind='bar', figsize=(8,5))
plt.locator_params(axis='x', nbins=10)
ax.set_ylabel('Number of Citations')
ax.set_title('Mean Citation by Year')
```

```python
ax = df.groupby('year').agg(['mean', 'median']).plot(kind='bar', figsize=(10,5))
plt.locator_params(axis='x', nbins=10)
ax.set_ylabel('Number of Citations')
ax.set_title('Average Citation by Year')
```

```python
df.n_citation.plot(kind='box')
```

```python
df_venue = df[['id', 'venue']].groupby('venue').count()
df_venue.sort_values('id', ascending=False).head(10).plot(kind='bar')
```
